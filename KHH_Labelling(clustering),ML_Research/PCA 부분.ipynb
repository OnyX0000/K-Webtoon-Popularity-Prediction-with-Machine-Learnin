```python
import pandas as pd
import ast
```

```python
final_df = pd.read_csv("../1월 프로젝트/final.csv")
```

```python
final_df.columns
```

```python
tag_df=pd.read_csv("../1월 프로젝트/웹툰 태그,인포.csv")
```

```python
tag_df=tag_df.drop('Unnamed: 0', axis=1)
tag_df
```

```python
tag_df.rename(columns={'웹툰 이름':'webtoon_name'},inplace=True)
tag_df
```

```python
columns= [x for x in list(final_df.columns) if "Tag" not in x and "Info" not in x] 
final_df=final_df[columns]
final_df
```

```python
tag_df=tag_df[['webtoon_name','tag']]
tag_df
```

```python
tag_df['tag']=tag_df['tag'].apply(ast.literal_eval)
```

```python
all_categories = set(item for sublist in tag_df['tag'] for item in sublist)
all_categories
```

```python
# 원핫 인코딩
all_categories = set(item for sublist in tag_df['tag'] for item in sublist)
for category in all_categories:
    tag_df[category] = tag_df['tag'].apply(lambda x: 1 if category in x else 0)
tag_df
```

```python
merge_final_df = pd.merge(final_df,tag_df,on='webtoon_name',how='inner')
merge_final_df
```

```python
merge_final_df=merge_final_df.drop_duplicates('webtoon_name')
merge_final_df.reset_index(drop=True,inplace=True)
merge_final_df
```

```python
merge_final_df.to_csv('final_final.csv',index=False)
```
